````python
# 1, 3, 4

def sep_tolist(data, col):
    data[col] = data[col].apply(lambda x: str(x).lower().replace('.', ','))
    data[col] = data[col].apply(lambda x: list(set(x.split(','))))

    L = data[col].values.tolist()
    LL = []
    for l in L:
        LL.extend(l)

    LL = set(LL)

    for c in LL:
        if c is not np.nan and c != 'nan':
            c = str(c).strip()
            data[f'{col}_{c}'] = data[col].apply(lambda x: int(c in x))

    data.drop(col, axis=1, inplace=True)
````

````python
def to_int(x):
    x = str(x)
    
    x = x.replace('OOR', '')
    x = x.replace('>', '').replace('<', '')
    x = x.replace(' м', '')
    
    if not x:
        return np.nan
    
    x = str(x).replace(',', '')
    
    try:
        return float(x)
    except Exception as e:
        #print(e)
        return x
````

````python
for col in ['gender', 'nationality', 'nation', 'genotype']:
    for c in data0[col].unique():
        if c is not np.nan and c != 'nan':
            c = str(c).strip()
            data0[f'{col}_{c}'] = data0[col].apply(lambda x: int(c is x))
    data0.drop(col, axis=1, inplace=True)

    
catcols = [
    'ECG', 'place_residence', 'psych_type', 'family_anamnesis', 'place_live',
    'hospitalization', 'disease'
]

for col in data0.select_dtypes(exclude=np.number):
    if len([str(s) for s in data0[col].unique() if str(s).count(',') > 1]) == 0 and col not in catcols:
        data0[col] = data0[col].apply(to_int)
    else:
        sep_tolist(data0, col)
````

````python
data1 = data1.rename(columns={'Protein ': 'PROTEIN_FFQ'})
data1['ID'] = data1['ID'].str.replace('С', 'C').values
data1['ID'] = data1['ID'].str.strip()

data0['Sample ID'] = data0['Sample ID'].str.replace('С', 'C').values



data = pd.merge(data0, data1, how='left', left_on='Sample ID', right_on='ID')
data.columns = data.columns.str.strip().str.upper()

data.to_excel('data_numeric.xlsx', index=False)
````

````python
def check_normality(data, name=''):
    
    fig, ax = plt.subplots(1, 2, figsize = (12, 4))
    ax[0].hist(data, edgecolor='k', bins='scott')
    ax[0].set_title('Histogram')
    scipy.stats.probplot(data, dist="norm", plot=plt, rvalue=True)
    ax[1].set_title('QQ-plot')

    shapiro = scipy.stats.shapiro(data)
    pvalue = np.round(shapiro.pvalue, 7)
    statistic = np.round(shapiro.statistic, 2)

    fig.suptitle(name + f' \n Shapiro-Wilk: statistic: {statistic}; pvalue: {pvalue}', y=1.1)


# K-S 
````

````python
from scipy import stats

var1, var2 = data[data.GROUP == 'CN'], data[data.GROUP != 'CN']



ttests = {}
ttests_all = {}


for col in both:
    
    if not var1[col].dropna().empty and not var2[col].dropna().empty:

        ttest = stats.mannwhitneyu(var1[col].dropna().values, var2[col].dropna().values)
        spearmanr1 = stats.spearmanr(data[col].dropna().values, data.dropna(subset=[col])['WEIGHT'].values, nan_policy='omit')
        spearmanr2 = stats.spearmanr(data[data.GROUP != 'CN'][col].dropna().values, data[data.GROUP != 'CN'].dropna(subset=[col])['AGE'].values, nan_policy='omit')
        
        pv2 = spearmanr1[1]
        pv3 = spearmanr2[1]
        pv = ttest.pvalue


        if not np.isnan(pv) and not np.isnan(pv2)and not np.isnan(pv3):

            ttests_all[col] = {'mannwhitneyu': ttest.statistic, 'mannwhitneyu-pval':ttest.pvalue,
                              'spearmanr-age-statistic': spearmanr2[0], 'spearmanr-age-pval': spearmanr2[1],
                              'spearmanr-weight-statistic': spearmanr1[0], 'spearmanr-weight-pval': spearmanr1[1]}
````

````python
pv = pd.DataFrame(ttests_all).T
````

````python
round(pv, 2).to_excel("part1.xlsx", index=True)
````

````python
pvs = pv[(pv['mannwhitneyu-pval'] < 0.05) & (pv['spearmanr-age-pval'] > 0.05)& (pv['spearmanr-weight-pval'] > 0.05)]
pvs
````

````python
pvs.index
````

````python
import seaborn as sns

top = list(pvs.index)

top = data[top].dropna()

for col in list(top):
    top[col] = top[col].apply('log')

top['group'] = data['GROUP']

sns.pairplot(top, hue='group', kind='reg')
````

## Adjust p-values !

````python
def fdr(p_vals):

    from scipy.stats import rankdata
    ranked_p_values = rankdata(p_vals)
    fdr = p_vals * len(p_vals) / ranked_p_values
    fdr[fdr > 1] = 1

    return fdr
````
