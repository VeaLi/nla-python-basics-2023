# Seaborn https://seaborn.pydata.org/


- `Seaborn` - это библиотека визуализации данных на языке `Python`, основанная на `matplotlib`. Она предоставляет высокоуровневый интерфейс для построения графиков. `Seaborn` в определенной степени может быть использоваться с `matplotlib`. Например, именование графиков или создание сетки для графиков проще сделать в `matplotlib`.

- В то время как библиотека `matplotlib` импортируется под псевдонимом `'plt'` командой `"import matplotlib.pyplot as plt"`, библиотека `seaborn` импортируется под псевдонимом `'sns'` командой `"import seaborn as sns"`. Чаще всего эти библиотеки импортируются вместе (последовательно на отдельных строках).

- И, поскольку для построения графиков, с помощью `seaborn` и `matplotlib`, наиболее часто нужны данные - с ними импортируют библиотеки `pandas` и `numpy`.

___

*Note:* Для импортирования библиотки `pandas` используется псевдонимом `'pd'` - командa `"import pandas as pd"`;
Для импортирования библиотки `numpy` используется псевдонимом `'np'` - командa `"import numpy as np"`.

## Попарная корреляция

Датафрейм `pandas` имеет множество встроенных функций. Например, `'.mean()'` - среднее значение, `'.std()'` - стандартная ошибка. A, метод `'.corr()'` возвращает матрицу корреляции признаков датафрейма; этот метод принимает любую функцию, которую он применяет попарно для каждой пары признаков. Метод `'.corr()'` также имеет встроенные функции `{'pearson', 'spearman', 'kendall'}`.

*Пример:*
```python
corr_spearman = data.corr('spearman')
print(corr_spearman)
```

### Рассчитайте корреляционную матрицу по методу Пирсона, сохраните ее в переменной `'corr'`.

*Note:* Каковы условия применения коэффициента корреляции Пирсона?
___

- Обе переменные находятся на одной шкале или интервале измерения;
- Переменные соответствуют нормальному распределению;
- В данных нет выбросов;
- Вы ожидаете, что переменные линейно зависят друг от друга.

````python
corr = data.corr('pearson')
corr
````

# Как создать график в Seaborn

Для рисования графиков в `seaborn` вам понадобиться поместить данные в датафрейм. Функции `seaborn` принимают датафрефм и аргументы:

*Пример:*
```python
sns.histplot(
    data=None, # Pandas DataFrame
    x=None, # Имя признака (название колонки) Pandas DataFrame
    y=None, # Имя признака (название колонки) Pandas DataFrame
    hue=None, # Имя признака (название колонки) - для разделить графиков по фактору, например  - полу
    stat='count',
    bins='auto',
    cumulative=False,
    kde=False,
    legend=True,
    ax=None,
)
```


## Гистограмма (Histogram)

- sns.histplot
- sns.distplot
- sns.kdeplot

### Aргументы `'x'` и `'y'`

````python
sns.histplot(data=data, x='sepal length (cm)');
````

````python
sns.histplot(data=data, y='sepal length (cm)');
````

### Aргумент `'hue'` (!)

````python
sns.histplot(data=data, x='sepal length (cm)', hue='species');
````

## Aргумент `'kde'` 

````python
sns.histplot(data=data, x='sepal length (cm)', hue='species', kde=True);

plt.savefig('histplot.pdf', bbox_inches = 'tight')
````

## Линейный график  и регрессия (Linear plot)

- sns.lineplot
- sns.regplot
- sns.lmplot

````python
sns.lineplot(data=data, x = 'sepal length (cm)', y = 'sepal width (cm)', hue='species')
````

````python
sns.lmplot(data=data, x = 'sepal length (cm)', y = 'sepal width (cm)', hue='species')
````

## Диграмма рассеивания (Scatter plot)

- sns.scatterplot

````python
sns.scatterplot(data=data, x = 'sepal length (cm)', y = 'sepal width (cm)', hue='species')
````

## Диграммы размаха

- sns.boxplot
- sns.boxenplot
- sns.violinplot

````python
sns.boxplot(data=data, y = 'sepal length (cm)', x = 'species');
````

````python
sns.boxenplot(data=data, y = 'sepal length (cm)', x = 'species');
````

## Столбчатая диаграмма (Barplot)

- sns.barplot

````python
sns.barplot(data=data, x='species', y = 'sepal length (cm)', capsize=0.2)
````

# Seaborn special: Heatmap, Clustermap, Relplot and Pairplot

- sns.heatmap
- sns.clustermap
- sns.pairplot
- [sns.relplot](https://seaborn.pydata.org/examples/heat_scatter.html)

````python
sns.heatmap(data=corr, cmap='RdYlBu', annot=True, fmt=".2f", alpha=0.9, linewidths=3, square=True);
````

````python
mask = np.triu(corr)
sns.heatmap(data=corr, cmap='RdYlBu', annot=True, fmt=".2f", alpha=0.9, linewidths=3, square=True, mask=mask);
````

````python
sns.pairplot(data=data)
````

````python
sns.pairplot(data=data, hue='species')
````

````python
sns.pairplot(data=data, kind='kde');
````

# sns.FacetGrid

Также можно разметить графики на сетке в seaborb, например, используя `sns.FacetGrid`:

````python
g = sns.FacetGrid(data, col="species")
g.map(sns.scatterplot, 'sepal length (cm)', 'sepal width (cm)');
````

# Matplotlib + seaborn

````python
sns.set_theme(style="whitegrid", palette="pastel", font_scale=1)

fig, axes = plt.subplot_mosaic("BRD;XXD", figsize=(12,8))

sns.scatterplot(data=data, x = 'sepal length (cm)', y = 'sepal width (cm)', hue='species', ax=axes['R'])
sns.barplot(data=data, x='species', y = 'sepal length (cm)', capsize=0.2, ax=axes['B'])
sns.boxenplot(data=data, y = 'sepal length (cm)', x = 'species', ax=axes['X']);
sns.histplot(data=data, y='sepal length (cm)', hue='species', kde=True, ax=axes['D']);

sns.despine()

fig.suptitle('$Iris$ dataset', size=22, y=0.95)
plt.savefig('matplotlib_with_seaborn.pdf', bbox_inches='tight')
````
